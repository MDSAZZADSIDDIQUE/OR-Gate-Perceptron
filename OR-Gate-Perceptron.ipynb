```python
import math
import matplotlib.pyplot as plt
import random
```

```python
X1 = [0, 0, 1, 1]
X2 = [0, 1, 0, 1]
Y  = [0, 1, 1, 1]
```

```python
plt.figure()
for i in range(len(Y)):
    if Y[i] == 0:
        plt.plot(X1[i], X2[i], "ro")
    else:
        plt.plot(X1[i], X2[i], "go")
plt.show()
```

```python
W = [random.random(), random.random()]
b = random.random()

print("Weights: ", W)
print("Bias: ", b)

learning_rate = 0.02
theta = 0.5
```

```python
epoch = 0
count = 0
loss = []
while count < 4:
    total_error = 0
    print("Epoch:", epoch)
    for i in range(len(X1)):
        weighted_sum = W[0] * X1[i] + W[1] * X2[i] + b
        pred = 0
        if weighted_sum > theta:
            pred = 1
        error = Y[i] - pred
        if error != 0:
            count = 0
            total_error += math.sqrt(error**2)
            W[0] += error * X1[i] * learning_rate
            W[1] += error * X2[i] * learning_rate
            b += error * learning_rate
        else:
            count += 1
        print(f"[{X1[i]}, {X2[i]}] = {pred}")
    epoch += 1
    print("Total loss: ", total_error)
    loss.append(1/4 * total_error)
```

```python
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
for i in range(len(Y)):
    if Y[i] == 1:
        plt.plot(X1[i], X2[i], "ro")
    else:
        plt.plot(X1[i], X2[i], "go")
plt.xlim([-1, 2])
plt.ylim([-1, 2])

xintr = (theta - b) / W[0]
yintr = (theta - b) / W[1]
slope = - yintr / xintr
xx = plt.xlim()
yeq = []
for x in xx:
    yeq.append(slope * x + yintr)
plt.plot(xx, yeq)
plt.subplot(1, 2, 2)
plt.plot(loss)
plt.show()
```
